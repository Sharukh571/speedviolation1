## checking for speed violation

```python
df1[df1.speed_mph>df1.speedlimit_mph]
```

## adding a new column as speed violation which shows 1 as speed violated and 0 as not violated

```python
row_indexes=df1[df1.speed_mph>df1.speedlimit_mph].index
```

```python
df1.loc[row_indexes,'speed_violation']="1"
```

```python
row_indexes=df1[df1.speed_mph<=df1.speedlimit_mph].index
```

```python
df1.loc[row_indexes,'speed_violation']="0"
```

```python
df1.head(50)
```

```python
df1.speed_violation.value_counts() ## speed violation counts
```

```python
## this graphs tells us that the maximum number of speed violation done by the drivers were mostly between 10 to 22 hours time.
pd.crosstab(df1.hour, df1.speed_violation).plot(kind='bar')
plt.show()
```

```python
## this graph tells us that the speed vioation was maximum on weekdays than weekends.
pd.crosstab(df1.Day_of_the_Week, df1.speed_violation).plot(kind='bar')
plt.show()
```

```python
pd.crosstab(df1.Month_Value, df1.speed_violation).plot(kind='bar')
plt.show()
```
