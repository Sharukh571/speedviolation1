## checking for missing values

```python
df.isnull().sum()
```

```python
## converting timestamp into datetime type
df['timestamp'] = pd.to_datetime(df['timestamp'])
df.info()
```

```python
## converting object type into integer type
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
df.controlledAccess = enc.fit_transform(df.controlledAccess)
df.state = enc.fit_transform(df.state)
```

## extracting month from timestamp and then adding new column as month

```python
df['timestamp'][0]
df['timestamp'][0].month
```

```python
list_months = []
list_months
```

```python
df.shape
```

```python
for i in range(df.shape[0]):
    list_months.append(df['timestamp'][i].month)
```

```python
list_months
```

```python
len(list_months)
```

```python
df['Month_Value'] = list_months
```

```python
df.head()
```

### extracting the day from timestamp and then adding a new column as week of the day

```python
df['timestamp'][2500].weekday()
```

```python
df['timestamp'][2500]
```

```python
def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
df['Day_of_the_Week'] = df['timestamp'].apply(date_to_weekday)
```

```python
df.head(100)
```

```python
df['timestamp'][100].time()
```

### extracting time and adding new column as hour

```python
df['hour'] = df['timestamp'].dt.hour
```

```python
df.head()
```

```python
## dropping timestamp column from data
df1 = df.drop('timestamp', axis=1)
df1.head()
```

```python
df1.corr()
```


## checking for outliers and then deleting

```python
from scipy import stats
from matplotlib import rcParams
```

```python
z =np.abs(stats.zscore(df1))
print(z)
```

```python
df1 = df1[(z<3).all(axis=1)]
```

```python
df1.shape
```

```python
rcParams['figure.figsize'] =10,9
df1.plot(kind='box')
```

## saving the new data as csv file

```python
df1.to_csv('vehicle_dataset.csv', index=False)
```
